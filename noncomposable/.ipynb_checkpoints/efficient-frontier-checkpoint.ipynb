```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp
```

This notebook provides code to plot the efficient frontier for the noncomposable hook problem.  Given a deterministic output function $G_1(\Delta)$ for a CFMM and a noncomposable hook with mean output $G_2(\Delta)$ and variance $\sigma^2(\Delta)$, we solve the problem: 

\begin{align*}
    \text{minimize} \quad & \sigma^2(\Delta) \\ 
    \text{subject to} \quad & G_1(\Delta) + G_2(100 - \Delta) \geq R,
\end{align*}
where $R > 0$ is a threshold return. We plot the efficient frontier, $R$ versus the optimal objective value. 

```python
# Define output functions or `mean returns'
def G1(x):
    return np.log(x + 1) 

def G2(x):
    return np.sqrt(x)
```

```python
# Define the functions G1 and G2 using CVXPY functions
def G1(x):
    return cp.log(x + 1)  # Example function, can be replaced

def G2(x):
    return cp.sqrt(x)  # Example function, can be replaced

# Define the variance functions
def variance_linear(delta, k):
    return k * delta

def variance_quadratic(delta, k):
    return k * delta**2

def variance_constant():
    return 3

# Simulation parameters
k = 10 #Set k = 10
target_returns = np.linspace(1, 10, 10)  # Range for R

efficient_frontier_linear = []
efficient_frontier_quadratic = []
efficient_frontier_constant = []

# Perform simulation for linear variance
for R in target_returns:
    delta = cp.Variable()
    objective = cp.Minimize(variance_linear(delta, k))
    constraints = [delta >= 0, delta <= 100, G1(100 - delta) + G2(delta) >= R]
    
    prob = cp.Problem(objective, constraints)
    result = prob.solve(qcp=True)
    
    efficient_frontier_linear.append(result)

# Perform simulation for quadratic variance
for R in target_returns:
    delta = cp.Variable()
    objective = cp.Minimize(variance_quadratic(delta, k))
    constraints = [delta >= 0, delta <= 100, G1(100 - delta) + G2(delta) >= R]
    
    prob = cp.Problem(objective, constraints)
    result = prob.solve(qcp=True)

    efficient_frontier_quadratic.append(result)

# Perform simulation for constant variance
sigma = variance_constant()
for R in target_returns:
    delta = cp.Variable()
    objective = cp.Minimize(sigma)
    constraints = [delta >= 0, delta <= 100, G1(100 - delta) + G2(delta) >= R]
    
    prob = cp.Problem(objective, constraints)
    result = prob.solve(qcp=True)

    efficient_frontier_constant.append(result)

# Plotting the results
plt.figure(figsize=(10, 6))

plt.plot(efficient_frontier_linear, target_returns, marker='o', color='blue', linestyle='-', label='Linear Variance')
plt.plot(efficient_frontier_quadratic, target_returns, marker='x', color='green', linestyle='--', label='Quadratic Variance')
plt.plot(efficient_frontier_constant, target_returns, marker='s', color='red', linestyle='-.', label='Constant Variance')

plt.xlabel('Optimal Variance', fontsize=12)
plt.ylabel('Target Return', fontsize=12)
plt.title('Efficient Frontier', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.savefig('figures/efficient_frontier.png', dpi=500)
plt.show()
```
