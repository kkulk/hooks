```python
import numpy as np
import matplotlib.pyplot as plt
```

This notebook plots the output functions $G_1(\Delta) = \log(\Delta + 1)$ and $G_2(\Delta) = \sqrt{\Delta}$ for the noncomposable hook simulations.  

```python
# Define output functions or `mean returns'
def G1(x):
    return np.log(x + 1) 

def G2(x):
    return np.sqrt(x)
```

```python
# Generate a range of Delta values
delta_values = np.linspace(0, 100, 500)

# Calculate G1 and G2
G1_values = G1(delta_values)
G2_values = G2(delta_values)

# Plotting the results
plt.figure(figsize=(10, 6))

plt.plot(delta_values, G1_values, label='$G_1(\\Delta) = \log(\\Delta + 1)$', color='blue', linestyle='-', linewidth=2)
plt.plot(delta_values, G2_values, label='$G_2(\\Delta) = \sqrt{\\Delta}$', color='green', linestyle='--', linewidth=2)

plt.xlabel('$\\Delta$', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.title('$G_1(\\Delta)$ and $G_2(\\Delta)$ vs $\\Delta$', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.savefig('mean-outputs.png', dpi=500)
plt.show()
```
